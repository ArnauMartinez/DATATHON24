```python
import pandas as pd

targets = pd.read_csv("targets.csv", sep=",", index_col=False)
product_data = pd.read_csv("product_data.csv", sep=",")
p_data = product_data[['cod_modelo_color', 'des_filename']]

targets.head()
```

```python
p_data.head()
```

```python


image_targets = pd.merge(p_data, targets, how='left')
image_targets = image_targets.drop(image_targets.columns[0], axis=1)

image_targets.head()
```

```python
image_targets.to_csv("images_targets.csv", index=False)
```
